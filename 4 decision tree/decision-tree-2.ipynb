```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/kaggle/input/social-network-ads/Social_Network_Ads.csv')
```

```python
# here all the starting columns can act as my input column & purchased column will act as target column
data.head()
```

```python
# convert gender to numerical col
data['Gender'].replace({'Male':0,'Female':1},inplace=True)
```

```python
x=data.iloc[:,1:4].values
y=data.iloc[:,-1].values
x.shape
```

```python
y.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier()
clf1.fit(x,y)
```

```python
from sklearn.tree import plot_tree
from matplotlib import pyplot as plt

plt.figure(figsize=(80, 50))  # Set plot size
plot_tree(clf1, filled=True)
plt.show()
```
